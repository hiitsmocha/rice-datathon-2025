## I. Explore each column independently

```python
def count_value_for_column(df, col_name):
    df_temp = df[[col_name, 'Vehicle Population']]
    distinct_values_count = df_temp[col_name].value_counts()
    population_sum = df_temp.groupby(col_name)['Vehicle Population'].sum()
    
    result_df = pd.DataFrame({
        'Distinct Values Count': distinct_values_count,
        'Population Sum': population_sum
    })
    return result_df
```

0. Date

```python
count_value_for_column(train_and_val, "Date")
```
