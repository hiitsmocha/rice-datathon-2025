# Import Required Libraries
Import the necessary libraries, including pandas and matplotlib.

```python
# Import the necessary libraries, including pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the Dataset
Load the dataset into a pandas DataFrame.

```python
# Load the Dataset
# Load the dataset into a pandas DataFrame
file_path_train = "/Users/huaijinsun/Downloads/training.xlsx"
df_train = pd.read_excel(file_path_train)
```

# Count Unknown/Blank Entries
Count the number of unknown/blank entries in each column of the DataFrame.

```python
# Count Unknown/Blank Entries
# Count the number of unknown/blank entries in each column of the DataFrame
missing_values_count = df_train.isnull().sum()
print('missing_value_count', missing_values_count)

# Plot the missing values count in a bar chart
plt.figure(figsize=(10, 6))
missing_values_count.plot(kind = 'bar')
plt.title('Count of Blank Entries in Each Column')
plt.xlabel('Columns')
plt.ylabel('Count of Unknown Entries')
plt.show()
```

```python
# Function to count unknown entries
def count_unknown_blank_entries(df):
    return df.isin(['Unknown', 'Not Applicable']).sum()

# Count unknown/blank entries in training dataset
unknown_blank_train = count_unknown_blank_entries(df_train)
print(unknown_blank_train)


# Plot bar chart for training dataset
plt.figure(figsize=(10, 5))
unknown_blank_train.plot(kind='bar', color='blue', alpha=0.7)
plt.title('Unknown/NA Entries in Training Dataset')
plt.xlabel('Columns')
plt.ylabel('Count')
plt.show()
```
