```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("scoring.csv")
```

```python
data
```

```python
df = pd.DataFrame(data)
df.drop_duplicates()
df
```

```python
print(df.dtypes)
```

```python
# Check for missing values in "Model Year"
missing_model_year = df["Model Year"].isna().sum()
print(f"Total missing 'Model Year' values: {missing_model_year}")
```

```python
df = df.dropna(subset=["Model Year"]).sort_values(by="Date", ascending=True).reset_index(drop=True)
```

```python
df["Model Year"] = df["Model Year"].astype(int)
```

```python
import json

# Generate summary JSON as a string
summary_json = df.describe(include="all").to_json()
summary_dict = json.loads(summary_json)
# Write the JSON dictionary to a file with indentation
with open("summary.json", "w") as f:
    json.dump(summary_dict, f, indent=4)
```

```python
df.drop(labels = "Region", axis = 1)
```

```python
df.to_csv("scoring_cleaned_data.csv", index=False)
```
